```python
import pandas as pd
import numpy as np
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram,linkage
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/content/drive/MyDrive/ISBR-Ravi/ML-2/Dataset/shopping_data.csv')
```

```python
X = data.iloc[:,3:].values
```

```python
model = linkage(X,method='ward')
```

```python
plt.figure(figsize=(12,8))
dendro = dendrogram(model)
plt.show()
```

```python
cluster = AgglomerativeClustering(n_clusters=5)
```

```python
cluster.fit_predict(X)
```

```python
plt.scatter(X[:,0],X[:,1],c=cluster.labels_)
plt.show()
```

```python
from skimage.data import coins
orig_coins = coins()
```

```python
import numpy as np
from scipy.ndimage import gaussian_filter
from skimage.transform import rescale

smoothened_coins = gaussian_filter(orig_coins, sigma=1)
rescaled_coins = rescale(
    smoothened_coins,
    0.2,
    mode="reflect",
    anti_aliasing=False,
)

#X = np.reshape(rescaled_coins, (-1, 1))
```

```python
plt.imshow(rescaled_coins,cmap='gray')
```
