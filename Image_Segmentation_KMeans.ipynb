```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
#reading the image
image = cv2.imread('images.jpg')
```

```python
#Convert the image from BGR to RGB
image= cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(image)
plt.show()
```

```python
image.shape
```

```python
#Reshaping to 2-D array from 3-D (RGB)
#image.reshape((-1,3)) -1 means original dimension of rows and columns will be multiplied
image=image.reshape((50421,3))
```

```python
image.shape
```

```python
#k-means algorithm (cv2) accepts the input to be in float32
image = np.float32(image)
```

```python
#The algorithm should stop if any of the exit criterion is reached
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER,200,0.95)
```

```python
compact,labels,center = cv2.kmeans(image,4,None,criteria,10,cv2.KMEANS_RANDOM_CENTERS)
```

```python
compact #WCSS or Inertia
```

```python
#Convert float values to 8-bits (gray scale values range from 0 to 255)
center = np.uint8(center)
```

```python
#Flatten the labels and subset
clustered_data = center[labels.flatten()]
```

```python
#Reshape to original shape
segmented_image = clustered_data.reshape((147, 343, 3))
```

```python
plt.imshow(segmented_image)
```
